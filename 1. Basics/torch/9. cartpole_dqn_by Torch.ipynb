# 강화학습 - 카트폴 게임 마스터하기

### DQN 학습 구조(Deep-q-network)
- 고전적 Q-Learning 강화학습 알고리즘 + 신경망 + a
- 강화 학습의 요소
    - State : 강화학습에서 학습하는 주체가 위치하거나 감지하고 있는 상태의 정보
    - Agent : 인공지능 플레이어
    - Environment : 에이전트가 솔루션을 찾기 위한 무대
    - Action : 에이전트가 환경 안에서 시행하는 상호작용
    - Reward 에이전트의 행동에 따른 점수 혹은 결과
- DQN의 특징: Memorize, Replay
- 카트폴 게임: 막대기가 넘어지지 않게 좌우 방향으로 학습 200이 되면 종료


```python
import gym
import random
import math
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque # Queue를 구현하기 위한
import matplotlib.pyplot as plt
```

### 하이퍼파라미터



```python
# 하이퍼파라미터
EPISODES = 50    # 애피소드 반복횟수(총 플레이할 게임 수)
EPS_START = 0.9  # 학습 시작시 에이전트가 무작위로 행동할 확률
EPS_END = 0.05   # 학습 막바지에 에이전트가 무작위로 행동할 확률
EPS_DECAY = 200  # 학습 진행시 에이전트가 무작위로 행동할 확률을 감소시키는 값
GAMMA = 0.8      # 할인계수
LR = 0.001       # 학습률
BATCH_SIZE = 64  # 배치 크기
```

## DQN 에이전트


```python
class DQNAgent:
    def __init__(self):
        self.model = nn.Sequential(
            nn.Linear(4, 256), # 카트 위치, 카트 속도, 막대기 각도, 막대기 속도 4가지 정보 입력 받음
            nn.ReLU(),
            nn.Linear(256, 2) # 왼쪽 갈때, 오른쪽 갈때 가치 출력
        )
        self.optimizer = optim.Adam(self.model.parameters(), LR)
        self.steps_done = 0 # 학습 반복할 때마다 증가
        self.memory = deque(maxlen=10000) # 큐가 가득찼을 때 먼저 들어온 것부터 지워줌
        # 기억된 경험은 다음 학습 시무작위로 뽑아 경험 간의 상관관계를 줄임
        
    def memorize(self, state, action, reward, next_state): 
        self.memory.append((state,
                            action,
                            torch.FloatTensor([reward]),
                            torch.FloatTensor([next_state])))
    
    def act(self, state):
        eps_threshold = EPS_END + (EPS_START - EPS_END) * math.exp(-1. * self.steps_done / EPS_DECAY)
        self.steps_done += 1
        if random.random() > eps_threshold: # threshold보다 크면 신경망이 학습하여 옳다고 생각하는 쪽으로 행동
            return self.model(state).data.max(1)[1].view(1, 1)
        else: # 작으면 무작위로 행동
            return torch.LongTensor([[random.randrange(2)]])
    
    def learn(self):
        if len(self.memory) < BATCH_SIZE: # 경험이 충분히 쌓이기 전까지는 기억 정리를 하지 않음
            return
        batch = random.sample(self.memory, BATCH_SIZE)
        states, actions, rewards, next_states = zip(*batch)

        states = torch.cat(states)
        actions = torch.cat(actions)
        rewards = torch.cat(rewards)
        next_states = torch.cat(next_states)

        current_q = self.model(states).gather(1, actions)
        max_next_q = self.model(next_states).detach().max(1)[0]
        expected_q = rewards + (GAMMA * max_next_q) # 행동들의 할인된 미래 가치
        
        loss = F.mse_loss(current_q.squeeze(), expected_q)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
```

## 학습 준비하기

`gym`을 이용하여 `CartPole-v0`환경을 준비하고 앞서 만들어둔 DQNAgent를 agent로 인스턴스화 합니다.

자, 이제 `agent` 객체를 이용하여 `CartPole-v0` 환경과 상호작용을 통해 게임을 배우도록 하겠습니다.

```python
env = gym.make('CartPole-v0')
agent = DQNAgent()
score_history = []
```

## 학습 시작

```python
for e in range(1, EPISODES+1):
    state = env.reset()
    steps = 0
    while True:
        env.render()
        state = torch.FloatTensor([state])
        action = agent.act(state)
        next_state, reward, done, _ = env.step(action.item())

        # 게임이 끝났을 경우 마이너스 보상주기
        if done:
            reward = -1

        agent.memorize(state, action, reward, next_state)
        agent.learn()

        state = next_state
        steps += 1

        if done:
            print("에피소드:{0} 점수: {1}".format(e, steps))
            score_history.append(steps)
            break
```

```python
plt.plot(score_history)
plt.ylabel('score')
plt.show()
```
